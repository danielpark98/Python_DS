상관분석

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
#상관분석 : 이상치, 선형성 전제를 확인해야댐 

tips = sns.load_dataset("tips")

tips.head()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot of tips")
plt.show()
```
