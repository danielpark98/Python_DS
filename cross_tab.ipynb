```python
import pandas as pd
df = pd.read_excel("./Data/Ecommerce/E_Commerce_Dataset.xlsx", sheet_name="Raw")
칼럼명 = ['CustomerID', 'Gender', 'MaritalStatus', 'CouponUsed', 'OrderCount']
df = df[칼럼명].copy()
df.head()
```

```python
#쿠폰 발행 <-> 매출 증가
pd.crosstab(df['CouponUsed'], df['OrderCount'], margins=True)

#쿠폰 발행하면 주문 횟수를 증가 시키긴 함
```

```python
#쿠폰 발행하면 구매자들의 몇퍼센트가 쿠폰발행의 영향을 받는지 분석하기 

(pd.crosstab(df['CouponUsed'], df['OrderCount'], margins=True).T \
    / pd.crosstab(df['CouponUsed'], df['OrderCount'], margins=True)['All']).T.iloc[:3,:]
```
